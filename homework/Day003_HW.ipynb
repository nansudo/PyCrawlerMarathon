### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict
with open('./data/64_72hr_CH.xml',encoding="utf-8") as fd:
    doc = dict(xmltodict.parse(fd.read()))
    
data =  doc['cwbopendata']['dataset']['locations']['location']   
length = len(data)

print('高雄有',length,'區的溫度資料')

for i in range(length):
    print(data[i]['locationName'])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

# for i in range(length):
#     print(data[i]['locationName'],' 時間:',data[0]['weatherElement'][0]['time'][0]['dataTime'],'溫度:',data[0]['weatherElement'][0]['time'][0]['elementValue']['value'])

for location in data:
    print(location['locationName'])
    print(location['weatherElement'][0]['time'][0]['dataTime'])
    print(location['weatherElement'][0]['time'][0]['elementValue'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

# print(data[0]['locationName'])

# weathLength = len(data[0]['weatherElement'][0]['time'])

# for i in range(weathLength):
#     print('時間:',data[0]['weatherElement'][0]['time'][i]['dataTime'], ' 溫度:',data[0]['weatherElement'][0]['time'][i]['elementValue']['value'])

location = data[0]
print(location['locationName'])
for l in location['weatherElement'][0]['time']:
    print(l['dataTime'] )
    print(l['elementValue'] )
```
