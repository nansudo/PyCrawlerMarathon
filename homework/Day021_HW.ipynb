## 作業目標

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime,timedelta
import time
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
timeRange = (datetime.now() - timedelta(1)).strftime('%Y/%m/%d')
# print(timeRange)


while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    if timeRange in soup.find(class_="part_list_2").findAll('span')[-1].text:
        print('結束')
        break


    
```

```python
for i in soup.find(class_="part_list_2").findAll('h3'):
    if timeRange in i.find('span').text:
        break
    print(i.find('span').text,i.find('em').text,i.find('a').text)

```

## 2. 取出現在時間兩小時內的新聞

```python
timeRang_2hr = (datetime.now() - timedelta(hours=2)).strftime('%Y/%m/%d %H:%M')

for i in soup.find(class_="part_list_2").findAll('h3'):
    if i.find('span').text < timeRang_2hr:
        break
    print(i.find('span').text,i.find('em').text,i.find('a').text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.support.ui import Select

timeRang_3d = (datetime.now() - timedelta(3)).strftime('%d')
timeRang_max =  (datetime.now() - timedelta(3)).strftime('%Y/%m/%d 17:00')
timeRang_min =  (datetime.now() - timedelta(3)).strftime('%Y/%m/%d 15:00')

s1 = Select(browser.find_element_by_id('selD'))
s1.select_by_value(timeRang_3d)

s2 = browser.find_element_by_id('button')
s2.click()
```

```python
while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    if timeRang_min > soup.find(class_="part_list_2").findAll('span')[-1].text:
        print('結束')
        break
```

```python
for i in soup.find(class_="part_list_2").findAll('h3'):
    if i.find('span').text < timeRang_min:
        break
    if i.find('span').text < timeRang_max:
        print(i.find('span').text,i.find('em').text,i.find('a').text)
```
