### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import time
import re

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

data = {
    'offset': 1,
    'tPage': 3,
    'tFile': '20200618.xml',
    'tOt': 0,
    'tSi': 100,
    'tAr': 0
}

status = True


for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
    

while status:
    res = requests.post('https://www.ettoday.net/show_roll.php',data)
    resSoup = BeautifulSoup(res.text, "html5lib")
    
    for d in resSoup.findAll('h3'):
        dataTime = d.find(class_="date").text
        if re.match('2020/06/18',dataTime):
            print(dataTime, d.find_all('a')[-1].text)
        else:
            status = False
    
    data['offset'] += 1
    time.sleep(1)    
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

title = {}

for i in soup.find(class_='part_menu_2').findAll('li'):
    if not i.find('a'):
        continue
    title.update({i.find('a').text: []})
    
print(title)
```

### ③ 哪一個類別的文章最多

```python
import requests
from bs4 import BeautifulSoup
import time
import re

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")


title = {}

for i in soup.find(class_='part_menu_2').findAll('li'):
    if not i.find('a'):
        continue
    title.update({i.find('a').text: []})


data = {
    'offset': 1,
    'tPage': 3,
    'tFile': '20200618.xml',
    'tOt': 0,
    'tSi': 100,
    'tAr': 0
}

status = True


for d in soup.find(class_="part_list_2").find_all('h3'):
    dataContent = d.find_all('a')[-1].text
#     print(d.find('em').text)
    title[d.find('em').text].append(dataContent)
    

while status:
    res = requests.post('https://www.ettoday.net/show_roll.php',data)
    resSoup = BeautifulSoup(res.text, "html5lib")
    
    for d in resSoup.findAll('h3'):
        dataTime = d.find(class_="date").text
        dataContent = d.find_all('a')[-1].text
        if re.match('2020/06/18',dataTime):
#             print(d.find('em').text)
            title[d.find('em').text].append(dataContent)
        else:
            status = False
    
    data['offset'] += 1
    time.sleep(1)

maxNum = 0    
maxName = ''

for i in title:
    if len(title[i]) > maxNum:
        maxNum = len(title[i])
        maxName = i
    
print('文章最多的類別:', maxName)
print('數量:',maxNum)
```
